```python
class Neural:
        
    def __init__(self):
        np.random.seed(10)
        self.wh = np.random.rand(7,10)
        self.wout = np.random.rand(7,8)
    
    def sigmoid(self, x, w):
        z = self.hypo(x,w)
        res = 1/(1 + np.exp(-z))
        return res
    
    def sigmoid_derivative(self, x, w):
        return self.sigmoid(x, w) * (1 - self.sigmoid(x, w))
    
    def hypo(self,x,w):
        res = 0
        for i in range(len(x)):
            res += x[i]*w[i]
        return res
    
    def fwdprop(self, inputs):
        z2 = []
        for i in range(7):
            z2.append(self.sigmoid(inputs, self.wh[i]))
        return z2
    
    def fwdpropout(self, inputs):
        inputs.insert(0,1)
        z3 = []
        for i in range(7):
            z3.append(self.sigmoid(inputs, self.wout[i]))
        return z3
    
    def finddeltaout(self, inputs, output):
        error = []
        if output == 1:
            for i in range(len(inputs)):
                if i == 0:
                    error.append(inputs[i] - 1)
                else:
                    error.append(inputs[i] - 0)
        
        if output == 2:
            for i in range(len(inputs)):
                if i == 1:
                    error.append(inputs[i] - 1)
                else:
                    error.append(inputs[i] - 0)
                    
        if output == 3:
            for i in range(len(inputs)):
                if i == 2:
                    error.append(inputs[i] - 1)
                else:
                    error.append(inputs[i] - 0)
                    
        if output == 4:
            for i in range(len(inputs)):
                if i == 3:
                    error.append(inputs[i] - 1)
                else:
                    error.append(inputs[i] - 0)
        
        if output == 5:
            for i in range(len(inputs)):
                if i == 4:
                    error.append(inputs[i] - 1)
                else:
                    error.append(inputs[i] - 0)
        
        if output == 6:
            for i in range(len(inputs)):
                if i == 5:
                    error.append(inputs[i] - 1)
                else:
                    error.append(inputs[i] - 0)
        
        if output == 7:
            for i in range(len(inputs)):
                if i == 6:
                    error.append(inputs[i] - 1)
                else:
                    error.append(inputs[i] - 0)
        return error
    
        
    def change_weighttoout(self, inputs, inputsh, errors):
        lr = 0.03
        for i in range(7):
            for j in range(8):
                if j == 7:
                    self.wout[i][j] = self.wout[i][j] - lr*errors[i]*(inputs[i]*(1-inputs[i]))
                    break
                self.wout[i][j] = self.wout[i][j] - lr*errors[i]*inputsh[j]*(inputs[i]*(1-inputs[i]))
            
    def change_weightin(self, errors, inputsh, inputso, inputs):
        lr = 0.03
        j = 0
        for i in range(7):
            for m in range(7):
                sum = 0
                sum = sum + (error[m] * (inputso[m]*(1-inputso[m])) * self.wout[i][m])
            for j in range(10):
                self.wh[i][j] = self.wh[i][j] - lr*inputs[j]*inputsh[i]*(1-inputsh[i])*sum
                
```

```python
network = Neural()
neterror = []
```
