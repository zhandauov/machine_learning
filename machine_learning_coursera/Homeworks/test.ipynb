```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
m = 0.3
b = 5.

def f(x):
    return m*x + b

def h(theta0, theta1, x):
    return theta0 + theta1*x
```

```python
X = np.random.randint(1000, size=50).astype(dtype=np.float32)
Y = np.array([f(i) + 10 * (random.random()-0.5) for i in X], dtype=np.float32)

theta0 = 0.0
theta1 = 0.0
lr = 0.000001
```

```python
plt.plot(X, Y, 'ro')
```

```python
for _ in range(1000):
    dt0 = 0
    dt1 = 0
    for i in range(len(X)):
        diff = h(theta0, theta1, X[i]) - Y[i]
        dt0 += diff
        dt1 += diff*X[i]

    dt0 /= len(X)
    dt1 /= len(X)

    theta0 -= lr*dt0
    theta1 -= lr*dt1

print(theta0, theta1)

_x = [0, 1000]
_y = [h(theta0, theta1, i) for i in _x]
```

```python
plt.plot(X, Y, 'ro')
plt.plot(_x, _y)
plt.show()
```
